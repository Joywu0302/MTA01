```python
distance = float(input("請輸入以公尺為單位的行駛距離"))
```

```python
distance_kms = distance / 1000 # 轉換為公里
time_minute = int (input("請輸入經過分鐘"))
time_sec = int (input("請輸入經過秒數"))
```

```python
time = time_minute*60 + time_sec
pace = time / distance_kms
print("步速是: " , str((pace//60)) + ":" + str((pace%60)))
```

```python
import yfinance
stock1=yfinance.Ticker('0050.TW')
df1=stock1.history(period='1y')
df1.head()
```

```python
df1.Close.plot(figsize=(20,10))
```

```python
stock2=yfinance.Ticker('0056.TW')
df2=stock2.history(period='1y')
df2.head()
df2.Close.plot(figsize=(20,10))
```

```python
import pandas as pd
df_many = pd.concat([df1.Close, df2.Close], axis=1)
df_many.columns=['0050','0056']
```

```python
df_many.head()
```

```python
df_many.plot()
```

```python
df_many['0050']=df_many['0050']/df_many.iloc[0][0]
df_many['0056']=df_many['0056']/df_many.iloc[0][1]
```

```python
df_many.head()
```

```python
df_many.plot()
```
